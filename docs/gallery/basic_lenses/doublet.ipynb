# Achromatic Doublet

```python
import numpy as np
from optiland import optic
```

Initialization empty 'Optic' object, which is the core class in Optiland representing an optical system.

```python
lens = optic.Optic()
```

Add surfaces, one at a time. We first define an BK7 lens, which is bonded to an SF2 lens. The lens radii and thicknesses are defined as follows.

```python
# add surfaces
lens.add_surface(index=0, radius=np.inf, thickness=np.inf)
lens.add_surface(index=1, radius=29.32908, thickness=0.7, material='BK7', is_stop=True)
lens.add_surface(index=2, radius=-20.06842, thickness=0.032)
lens.add_surface(index=3, radius=-20.08770, thickness=0.5780, material=('SF2', 'schott'))
lens.add_surface(index=4, radius=-66.54774, thickness=47.3562)
lens.add_surface(index=5)
```

Add the aperture. We choose image-space F-number (imageFNO) as the aperture type and choose a value of 8 mm.

```python
# add aperture
lens.set_aperture(aperture_type='imageFNO', value=8.0)
```

Set the field type to angle and choose fields at 0.0, 0.7 and 1.0 degrees.

```python
# add field
lens.set_field_type(field_type='angle')
lens.add_field(y=0.0)
lens.add_field(y=0.7)
lens.add_field(y=1.0)
```

We add wavelengths across the visible spectrum based on the standard F, d and C lines Fraunhofer lines. The wavelength at 0.5876 microns is defined as the primary wavelength.

```python
# add wavelength
lens.add_wavelength(value=0.4861)
lens.add_wavelength(value=0.5876, is_primary=True)
lens.add_wavelength(value=0.6563)
```

View the system in 2D:

```python
lens.draw(num_rays=10)
```
